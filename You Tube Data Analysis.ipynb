## Top 10 You Tube Channels by Subscribers

```python
top_channels = df.sort_values(by = 'subscribers',ascending=False).head(10)
plt.bar(top_channels['Youtuber'],top_channels['subscribers'])
plt.title("Top 10 You Tube Channels By Subscribers")
plt.xlabel("Channel Name")
plt.ylabel("Subscribers")
plt.xticks(rotation=90)
plt.show()
```

## Insights
### Top Channels millions ya billions ke sath dominate karte hai
### Entertainment aur music channels usually top pe hote hai

# Category Wise You Tube Channels

```python
category_count = df['category'].value_counts().head(10)
category_count.plot(kind='bar')
plt.title("Category Wise You Tube Channels")
plt.xlabel("You Tube Channels")
plt.ylabel("Category")
plt.show()
```

## Insights
### Entertainment aur Music category dominent pe hai 
### Gaming aur educational channels bhi popular hote ja rahe hai 
### Comedy aur animation bhi thik thak condition me chal rahi hai

# Top Countries with most You Tube Channels 

```python
df.columns
```

```python
country_count = df['Country'].value_counts().head(10)
country_count.plot(kind ='bar')
plt.title("Top Countries With most You Tube Channels")
plt.xlabel("Country")
plt.ylabel("Number of Channels")
plt.show()
```

# Subscribers Vs video views

```python
plt.scatter(df["subscribers"],df["video views"])
plt.title("Subscribers Vs Video Views")
plt.xlabel('Subscribers')
plt.ylabel('Video Views')
plt.show()
```

## Insights 
### there is a strong positive trend betwen subscribers and video views
### Majority creators are small channels
### Some very large channel dominate the platforms
### Views depend not only on Subscribers,Content strategy also very importent for video views
